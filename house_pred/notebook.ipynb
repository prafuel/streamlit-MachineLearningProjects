```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
from sklearn.linear_model import LinearRegression
```

MODEL training only using numerical features

```python
numerical = np.array(df[['lotsize','bedrooms','bathrms','stories','garagepl']])
```

```python
target = np.array(df['price'])
```

```python
model1 = LinearRegression()
```

```python
X_train, x_test, y_train, y_test = train_test_split(numerical, target, test_size=0.3, random_state=555)
```

```python
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(x_test)
```

```python
print(r2_score(y_test, y_pred))
```

```python
score = []
for i in range(1000):
    X_train, x_test, y_train, y_test = train_test_split(numerical, target, test_size=0.3, random_state=i)
    model1.fit(X_train, y_train)
    y_pred = model1.predict(x_test)
    print(i,r2_score(y_test, y_pred))
    score.append((r2_score(y_test, y_pred), i))
```

```python
max(score)
```

```python
X_train, x_test, y_train, y_test = train_test_split(numerical, target, test_size=0.3, random_state=max(score)[1])
model1.fit(X_train, y_train)
y_pred = model1.predict(x_test)
print(r2_score(y_test, y_pred))
```

Training with all Columns

```python
df
```

```python
def change_fun(string : str):
    if string == "yes":
        return 1
    return 0

for column in df.columns:
    if len(pd.get_dummies(df[column]).columns) == 2:
        df[column] = df[column].apply(change_fun)
```

```python
X = np.array(df.drop(columns=['price']))
y = np.array(df['price'])
```

```python
df.to_csv("clean_house.csv")
```
